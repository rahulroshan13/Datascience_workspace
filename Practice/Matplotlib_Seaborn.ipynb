```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np

a= np.arange(1,11)
```

```python
a
```

```python
b = a**2
```

```python
b
```

```python
plt.plot(a,b)
```

```python
plt.plot(a,b, 'r')
plt.xlabel('x label title')
plt.ylabel('y label')
plt.title('title of the plot')
```

```python
plt.subplot(1,2,1)
plt.plot(a,b,'r*-')
plt.subplot(1,2,2)
plt.plot(b,a,'bo-')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes.set_xlabel('x-axis')
axes.set_ylabel('y-axis')
axes.set_title('title of axes')
axes.plot(a,b)
axes2.set_xlabel('x-axis')
axes2.set_ylabel('y-axis')
axes2.set_title('title of axes')
axes2.plot(b,a)
```

```python
print(type(fig))
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,0.35,0.8])
axes2 = fig.add_axes([0.45,0,0.35,0.8])
axes.plot(a,b)
axes2.plot(b,a)
```

```python
fig.savefig('myfig.png')
```
