```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
sns.countplot(x = 'sex', data=df)
```

```python

sns.countplot(x = 'time', data = df)
```

```python
sns.countplot(x = 'time', data = df, hue='smoker')
```

```python
sns.barplot(x = 'sex', y = 'total_bill', data = df)
```

```python
df.groupby('sex')['total_bill'].mean()
```

```python
sns.barplot(x = 'sex', y = 'total_bill', data = df, hue = 'time')
```

```python
df.groupby('time').mean()
```

```python
sns.boxplot(x = 'sex', y = 'total_bill', data = df)
```

```python
sns.boxplot(x = 'sex', y = 'total_bill', data = df, hue = 'time')
```

```python
sns.boxplot(x = 'time', y = 'tip', data = df, hue = 'size')
```

```python
sns.violinplot(x = 'day', y ='total_bill', data = df)
```

```python
sns.violinplot(x = 'day', y ='total_bill', data = df, hue = 'sex')
```

```python
sns.violinplot(x = 'day', y ='total_bill', data = df, hue = 'sex', split=True)
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = df, jitter=False)
```
